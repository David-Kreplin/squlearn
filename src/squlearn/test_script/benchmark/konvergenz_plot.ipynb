```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt
import pandas as pd

def plot_convergence(file_path):
    """
    Plots convergence based on values from a file.

    Parameters:
    - file_path: Path to the file containing convergence data.

    Returns:
    - None
    """
    # Lese die Daten aus der Datei
    data = pd.read_csv(file_path, delim_whitespace=True)

    # Erstelle den Konvergenzplot
    plt.figure(figsize=(10, 6))
    plt.plot(data['Iteration'], data['f(x)'], marker='o', linestyle='-')
    plt.title('Convergence Plot')
    plt.xlabel('Iteration')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()

# Beispielaufruf
plot_convergence('plot')
```
