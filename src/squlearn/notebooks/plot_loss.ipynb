```python
import matplotlib.pyplot as plt

# Laden Sie die Log-Datei und extrahieren Sie die relevanten Informationen
iteration = []
loss = []

with open("plot", "r") as file:
    next(file)  # Überspringen Sie die Header-Zeile
    for line in file:
        parts = line.split()
        iteration.append(int(parts[0]))
        loss.append(float(parts[1]))

# Erstellen Sie den Optimierungsverlauf-Plot
plt.figure(figsize=(10, 6))
plt.plot(iteration, loss, linestyle='-', color='b')
plt.yscale("log")
plt.title('Optimierungsverlauf(Adam)')
plt.xlabel('Iteration')
plt.ylabel('Funktionswert (f(x))')
plt.grid(True)
```
