```python
from squlearn.feature_map import ChebPQC
from squlearn.feature_map.feature_map_derivatives_optree import FeatureMapDerivatives
from squlearn.feature_map.feature_map_derivatives import FeatureMapDerivatives as FeatureMapDerivatives_old
```

```python
fm = ChebPQC(6,2,2)
fm.draw()
```

```python
fmd = FeatureMapDerivatives(fm)
print(fmd.get_derivative("laplace"))
```

```python
fmd_old = FeatureMapDerivatives_old(fm)
print(fmd_old.get_derivative("laplace"))
```
