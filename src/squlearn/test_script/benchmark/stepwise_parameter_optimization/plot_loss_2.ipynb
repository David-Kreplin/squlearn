```python
import matplotlib.pyplot as plt

# Laden Sie die Log-Datei und extrahieren Sie die relevanten Informationen fr Adam
adam_iteration = []
adam_loss = []

with open("adam_params_9_2.log", "r") as adam_file:
    next(adam_file)  # berspringen Sie die Header-Zeile
    for line in adam_file:
        parts = line.split()
        adam_iteration.append(int(parts[0]))
        adam_loss.append(float(parts[1]))

# Laden Sie die Log-Datei und extrahieren Sie die relevanten Informationen fr SGLBO
sglbo_iteration = []
sglbo_loss = []

with open("sglbo_params_9_2.log", "r") as sglbo_file:
    for _ in range(8):      # berspringen Sie die Header-Zeile
        next(sglbo_file) 
        
    for line in sglbo_file:
        parts = line.split()
        sglbo_iteration.append(int(parts[0]))
        sglbo_loss.append(float(parts[1]))
        

# Erstellen Sie den Optimierungsverlauf-Plot
plt.figure(figsize=(15, 9))

# Plot fr Adam
plt.plot(adam_iteration, adam_loss, linestyle='-', color='#729ECE', label='Adam', linewidth=4)

# Plot fr SGLBO
plt.plot(sglbo_iteration, sglbo_loss, linestyle='-', color='#67BF5C', label='SGLBO', linewidth=4)

plt.yscale("log")

plt.title(r'$f(x) = sin(6x), \ \theta = 9$', fontsize=34)

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.xlabel("Iteration", fontsize=28)
plt.ylabel("Squared Loss", fontsize=28)

plt.grid(True)

plt.legend(fontsize=28)  # Fgt die Legende hinzu
```

```python
# Erstellen Sie den Optimierungsverlauf-Plot
plt.figure(figsize=(15, 9))

# Plot fr Adam
plt.plot(adam_iteration, adam_loss, linestyle='-', color='b', label='Adam', linewidth=2)

# Plot fr SGLBO
plt.plot(sglbo_iteration, sglbo_loss, linestyle='-', color='r', label='SGLBO', linewidth=2)

plt.yscale("log")

plt.title(r'$f(x) = sin(6x), \ \theta = 9$', fontsize=18)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.xlabel("Iteration", fontsize=16)
plt.ylabel(r'$f_{\theta}(x)$', fontsize=16)

plt.grid(True)

#plt.legend(fontsize=16)  # Fgt die Legende hinzu
```
