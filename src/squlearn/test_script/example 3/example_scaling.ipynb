```python
import numpy as np
import matplotlib.pyplot as plt


def square(x,y):
    return x**2 + y**2

def square_grad(x,y):
    return np.array([2*x, 2*y])
```

```python
# Lese die Optimierungskoordinaten aus der Datei "params.log"
plt.figure(figsize=(10,10))

# plot the decision function for each datapoint on the grid
xx, yy = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = []
for i in range(len(xx)):
    for j in range(len(xx[i])):
        param = np.array([xx[i][j], yy[i][j]])
        Z.append(square(param[0], param[1]))

Z = np.array(Z)
Z = Z.reshape(xx.shape)
#norm = mpl.colors.Normalize(vmin=-max(abs(np.min(Z)),abs(np.max(Z))), vmax=max(abs(np.min(Z)),abs(np.max(Z))))
im = plt.imshow(
    Z,
    interpolation="bilinear",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    aspect="auto",
    origin="lower",
    cmap=plt.cm.PuOr_r,
    #norm=norm
)

plt.axis([-3, 3, -3, 3])


# Plot backtransformed hyperplane
contours = plt.contour(xx, yy, Z, levels=[0,1,2,3,4], linewidths=2, linestyles="dashed")

plt.plot(1,1,'x',color='black',markersize=10,linewidth=2)
plt.plot([1,1+square_grad(1,1)[0]],[1,1+square_grad(1,1)[1]],'-',color='black',markersize=10,linewidth=2)

# Plot training data

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Contour Plot with Optimization Path')
plt.legend()
plt.show()
```
