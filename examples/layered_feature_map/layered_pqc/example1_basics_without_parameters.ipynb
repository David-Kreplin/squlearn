```python
from squlearn.feature_map.layered_feature_map import LayeredPQC,VariableGroup,LayerPQC
import numpy as np
from qiskit.circuit import ParameterVector
```

```python
"""Initializing a feature map with basic non-parameter operations"""
# Initialize a feature map without parameters with the following code; with number_of_qubits=<number> you set the number of qubits used in this feature map: (example with 3 qubits)
feature_map_without_param = LayeredPQC(num_qubits=3)

# Note that you can already initialize gate layers, which do not require parameters:

# -----------------------------------------------------------------------------------------------------------------------------------
# Single qubit operations without parameters:
feature_map_without_param.H()
feature_map_without_param.X()

feature_map_without_param.Z()
feature_map_without_param.Y()
feature_map_without_param.I()
feature_map_without_param.S()
feature_map_without_param.S_conjugate()
feature_map_without_param.T()
feature_map_without_param.T_conjugate()
# -----------------------------------------------------------------------------------------------------------------------------------


# To print the feature map write the following code
feature_map_without_param.get_circuit().draw()
```

```python
# -----------------------------------------------------------------------------------------------------------------------------------
# Two qubit operations without parameters:
# You can choose between nearest neighbour ("NN") or all in all ("AA") entangling in every two qubit operation ("NN" is default):

feature_map_without_param = LayeredPQC(num_qubits=3)

feature_map_without_param.ch_entangling()
feature_map_without_param.ch_entangling(ent_strategy="NN")
feature_map_without_param.ch_entangling(ent_strategy="AA")
feature_map_without_param.cx_entangling()
feature_map_without_param.cy_entangling()
feature_map_without_param.cz_entangling()
feature_map_without_param.swap()
feature_map_without_param.swap(ent_strategy="AA")
# -----------------------------------------------------------------------------------------------------------------------------------


feature_map_without_param.get_circuit().draw()
```
